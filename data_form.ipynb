```python
# df.drop(columns = [license], inplace= True)
list(df.columns)
df.drop(columns = ['license'], inplace= True)
```

```python
#Display the country code 
country_code = df['country code']
print(country_code)
```

```python
#Chuan hoa ten cot (Column name standardization, lower case and replace space with underscore)
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

df.columns
```

```python
hosts_df = df[[
    "host_id",
    "host_name",
    "host_identity_verified",
    "calculated_host_listings_count"
]].drop_duplicates()
hosts_df.to_csv("db-airbnb/hosts.csv", index=False)
hosts_df.head()
```

```python
locations_df = df[[
    "neighbourhood_group",
    "neighbourhood",
    "lat",
    "long",
    "country",
    "country_code"
]].drop_duplicates().reset_index(drop=True)
locations_df.insert(0, "location_id", locations_df.index + 1)
locations_df.to_csv("db-airbnb/locations.csv", index=False)
locations_df.head()
```

```python
df = df.merge(
    locations_df,
    on=["neighbourhood_group", "neighbourhood", "lat", "long", "country", "country_code"],
    how="left"
)
df.head()
```

```python
rooms_df = df[[
    "id",
    "name",
    "host_id",
    "location_id",
    "room_type",
    "construction_year",
    "minimum_nights",
    "availability_365",
    "house_rules",
    "cancellation_policy",
    "instant_bookable"
]].drop_duplicates()
rooms_df = rooms_df.rename(columns={"id": "room_id"})
rooms_df["construction_year"] = (
    rooms_df["construction_year"]
    .astype(float)
    .round()
    .astype("Int64")
)
rooms_df["minimum_nights"] = (
    rooms_df["minimum_nights"]
    .astype(float)
    .round()
    .astype("Int64")
)
rooms_df["availability_365"] = (
    rooms_df["availability_365"]
    .astype(float)
    .round()
    .astype("Int64")
)

rooms_df.to_csv("db-airbnb/rooms.csv", index=False)
rooms_df.head()
```

```python
review_stats_df = df[[
    "id",
    "number_of_reviews",
    "last_review",
    "reviews_per_month",
    "review_rate_number"
]].drop_duplicates()

review_stats_df = review_stats_df.rename(columns={"id": "room_id"})

# ===== CLEAN DTYPE =====

# room_id: int
review_stats_df["room_id"] = review_stats_df["room_id"].astype("Int64")

# number_of_reviews: float -> int
review_stats_df["number_of_reviews"] = (
    review_stats_df["number_of_reviews"]
    .astype(float)
    .round()
    .astype("Int64")
)

# last_review: object -> date
review_stats_df["last_review"] = (
    pd.to_datetime(review_stats_df["last_review"], errors="coerce")
    .dt.date
)

# reviews_per_month: float -> 2 decimals
review_stats_df["reviews_per_month"] = (
    review_stats_df["reviews_per_month"]
    .astype(float)
    .round(2)
)

# review_rate_number: float -> 2 decimals
review_stats_df["review_rate_number"] = (
    review_stats_df["review_rate_number"]
    .astype(float)
    .round(2)
)

# ===== EXPORT =====
review_stats_df.to_csv("db-airbnb/review_stats.csv", index=False)

review_stats_df.head()
```

```python
pricing_df = df[[
    "id",
    "price",
    "service_fee"
]].drop_duplicates()
pricing_df = pricing_df.rename(columns={"id": "room_id"})
pricing_df.insert(0, "pricing_id", range(1, len(pricing_df) + 1))
pricing_df["effective_from"] = pd.to_datetime("today").normalize()
pricing_df["price"] = (
    pricing_df["price"]
      .str.replace("$", "", regex=False)
      .str.replace(",", "", regex=False)
)
pricing_df["service_fee"] = (
    pricing_df["service_fee"]
      .str.replace("$", "", regex=False)
      .str.replace(",", "", regex=False)
)
pricing_df.to_csv("db-airbnb/pricing.csv", index=False)
```

```python
hosts_df = df[[
    "host_id",
    "host_name",
    "host_identity_verified",
    "calculated_host_listings_count"
]].drop_duplicates()
# hosts_df.to_csv("db-airbnb/hosts.csv", index=False)
# # Fix boolean
hosts_df["host_identity_verified"] = hosts_df["host_identity_verified"].str.lower().map({
    "verified": True,
    "unconfirmed": False,
    "true": True,
    "false": False
})
# # Fix int column
hosts_df["calculated_host_listings_count"] = (
    hosts_df["calculated_host_listings_count"]
    .fillna(0)
    .astype(int)
)
hosts_df.to_csv("db-airbnb/hosts.csv", index=False)
```
